```python
import pandas as pd
import os

os.chdir('C:/Users/p_adi')
# Load the patient data from the CSV file
patient_data = pd.read_csv("relapse_kipu.csv")
patient_data.head()
```

```python


# Define a function to apply the label to each diagnosis code
def label_diagnosis_code(diagcode):
    if diagcode.startswith(('F20.', 'F31.', 'F32.3', 'F33.', 'F41.1', 'F42.', 'F43.21')):
        return 'severe'
    elif diagcode.startswith(('F40.', 'F43.0', 'F43.10', 'F44.', 'F48.0')):
        return 'less severe'
    else:
        return 'other'

# Apply the label to each diagnosis code
patient_data['diag_label'] = patient_data['diagcode'].apply(label_diagnosis_code)

# Group the data by patient ID and concatenate the labels for each record
grouped_data = patient_data.groupby('patientmasterkey')['diag_label'].apply(lambda x: ','.join(x)).reset_index()

# Split the concatenated labels and count the number of occurrences of each label for each patient
grouped_data['label_list'] = grouped_data['diag_label'].str.split(',')
grouped_data = grouped_data.explode('label_list')
grouped_data = grouped_data.groupby(['patientmasterkey', 'label_list']).size().reset_index(name='count')

# Find the label with the maximum count for each patient
grouped_data = grouped_data.sort_values(['patientmasterkey', 'count'], ascending=[True, False])
grouped_data = grouped_data.groupby('patientmasterkey').first().reset_index()

# Merge the data back with the original patient data
result = pd.merge(patient_data.drop(columns='diag_label'), grouped_data, on='patientmasterkey')
```
